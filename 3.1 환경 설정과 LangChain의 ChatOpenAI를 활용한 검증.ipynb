# 3. LLM 답변 생성

- OpenAI 대시보드에서 발급받은 API Key를 `OPENAI_API_KEY`라고 저장하면 별도의 설정 없이 `ChatOpenAI`를 사용할 수 있음

```python
from langchain_openai import ChatOpenAI

llm_openai = ChatOpenAI(model="gpt-4o-mini")
```

```python
question = "인프런에 어떤 강의가 있나요?"
llm_openai_message = llm_openai.invoke(question)
```

```python
print(llm_openai_message.content)
```

```python
from langchain_ollama import ChatOllama

question = "인프런에 어떤 강의가 있나요?"
llm_ollama = ChatOllama(model="qwen2.5")
llm_ollama_message = llm_ollama.invoke(question)
print(llm_ollama_message.content)
```
